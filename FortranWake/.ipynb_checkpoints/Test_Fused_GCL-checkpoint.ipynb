If the power and trhust coefficient curves linear interpolation needs to be improved then the p_c and ct_c properties of the  wt_layout instance can be updated

```python
nu = 100
p_c = np.array([[[np.linspace(wt_layout.wt_array(attr='cut_in_wind_speed')[j],
                              wt_layout.wt_array(attr='cut_out_wind_speed')[j],nu)[i],
                  interp(wt_layout.wt_array(attr='power_curve')[j][:,0],
                         wt_layout.wt_array(attr='power_curve')[j][:,1],
                         np.linspace(wt_layout.wt_array(attr='cut_in_wind_speed')[j],
                                     wt_layout.wt_array(attr='cut_out_wind_speed')[j],nu)[i])] \
                 for i in range(nu)] for j in range(wt_layout.n_wt)])

ct_c = np.array([[[np.linspace(wt_layout.wt_array(attr='cut_in_wind_speed')[j],
                              wt_layout.wt_array(attr='cut_out_wind_speed')[j],nu)[i],
                  interp(wt_layout.wt_array(attr='c_t_curve')[j][:,0],
                         wt_layout.wt_array(attr='c_t_curve')[j][:,1],
                         np.linspace(wt_layout.wt_array(attr='cut_in_wind_speed')[j],
                                     wt_layout.wt_array(attr='cut_out_wind_speed')[j],nu)[i])] \
                 for i in range(nu)] for j in range(wt_layout.n_wt)])
```

```python
pos=wt_layout.wt_array(attr='position')
# Compute the turbine to turbine vector in global coordinates
x_g = np.zeros([wt_layout.n_wt,wt_layout.n_wt])
y_g = np.zeros([wt_layout.n_wt,wt_layout.n_wt])
z_g = np.zeros([wt_layout.n_wt,wt_layout.n_wt])
for i in range(wt_layout.n_wt):
    for j in range(wt_layout.n_wt):
        x_g[i,j] = pos[j,0]-pos[i,0]
        y_g[i,j] = pos[j,1]-pos[i,1]
        z_g[i,j] = wt_layout.wt_array(attr='hub_height')[j] \
                 - wt_layout.wt_array(attr='hub_height')[i]
```

```python
nWS = len(wind_speeds)
nWD = len(wind_directions)
WS,WD = np.meshgrid(np.array(wind_speeds),np.array(wind_directions))
WS = WS.flatten()
WD = WD.flatten()
```

```python
def shear_u_eq_correction(H,R,
                          alpha=0.14,
                          order=2.,
                          NG=10,
                          N_sample=100000):
    '''
     order = 3 for power
     order = 2 for thrust
     order = 1 for average
    '''
    r_Gc,w_Gc = np.polynomial.legendre.leggauss(NG)
    wj,wk=np.meshgrid(w_Gc,w_Gc)
    tj,rk=np.meshgrid(r_Gc,r_Gc)

    z_ref  = H
    z_eval = z_ref + R*(rk+1.)*np.sin(np.pi*(tj+1.))/2.0

    int_a = 0.25*np.pi*R**2.*\
            np.sum( wj*wk*( (z_eval/z_ref)**(order*alpha) )*(rk+1.))
    int_ref = 0.25*np.pi*R**2.*\
            np.sum( wj*wk*( (z_eval/z_ref)**(order*0.) )*(rk+1.))

    factor = (int_a/int_ref)**(1.0/order)

    return factor

```

```python
ws_corr_factor= shear_u_eq_correction(H=wt_layout.wt_array(attr='hub_height')[0], 
                      R=wt_layout.wt_array(attr='rotor_diameter')[0]/2., 
                      alpha=0.14,order=3.,NG=4)
ws_corr_factor
```

```python
WF = MultipleFusedFGCL()
WF.wind_speeds=[wind_speeds[i]*ws_corr_factor for i in range(len(wind_speeds))]
WF.wind_directions=wind_directions 
WF.wt_layout = wt_layout
WF.run()
WF.Power/1e6
```
